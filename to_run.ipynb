# SEIRD-compartmental-agent-based-model
This code written in Python2 simulate an SEIRD (Susceptible-Exposed-Infected-Recovered or Dead) model 
with immunization process on a Scale Free network, built with the Barab√°si-Albert model with nodes as agents.

Varying the parameters of the model different epidemic scenarios can be explored and the effect of a random immunization 
process can be tested.

Each node can change its state due to a spontaneous transition which occours after a cretain time or due to an 
interaction with an infected node.

All the parameters are inserted by the user at the beginning of the simulation and state of the nodes and epidemic progression
is shown.


# Without running the %pylab inline command, it won't be able to show the plots.

# Here a guideline to choose the parameters of the model:

All the parameters (infection rate, recovery rate, exposed rate, hospitalized-to-recovered rate, hospitalized-to-dead rate, mortality rate, immunization rate) must be between 0 and 1. 
Recovery rate, exposed rate, hospitalized-to-recovered rate, hospitalized-to-dead rate are all inverse of duration periods, once that period is chosen the related parameter is obtained (ex. 10 day needed be hospitalized after infection, then the recovery rate is 0.1).

Values to test the model on a 1000 nodes network was: 

Infection rate [0.5,0.7]

Recovery rate [0.1,0.3]

exposed rate [0.1,0.4]

hospitalized-to-recovered rate [0.025,0.05]

hospitalized-to-dead rate [0.01,0.03]

mortality rate [0.05,0.2]

immunization rate [0.05,0.2]
